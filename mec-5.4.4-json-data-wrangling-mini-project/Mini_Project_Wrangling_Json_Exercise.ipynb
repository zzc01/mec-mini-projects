## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
# In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
bank_data = json.load(open('data/world_bank_projects.json'))
print(type(bank_data), len(bank_data))
# bank_data[1]
```

```python
# Find the top 10 major project themes (using column 'mjtheme_namecode')
# In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.
bank_df = json_normalize(bank_data, 'mjtheme_namecode')
bank_df.head(5)
# json_normalize(bank_data, 'mjtheme')
```

```python
bank_df[bank_df['code']=='1'][:10]
```

```python
bank_df.groupby('code').size().sort_values(ascending=False)[:10]
```

### 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Here can see in the name column there is empty string 
bank_sort_df = bank_df.sort_values(['code','name'], ascending=False)
bank_sort_df[bank_sort_df['code']=='1']
```

```python
# This is to check if a group of code have empty name 
bank_sort_df.groupby('code').max()
```

```python
# There are null values in the name column 
import numpy as np
bank_sort_fill_df = bank_sort_df.replace(to_replace='', value=np.nan)
bank_sort_fill_df.info()
```

```python
# After filling the nan, there is no null values 
bank_sort_fill_df.fillna(method='ffill', inplace=True)
bank_sort_fill_df.info()
```

```python
# Check if the result matches with question 2.
bank_sort_fill_df.groupby(['code','name']).size().sort_values(ascending=False)
```

```python
# Here is the dataframe with the name filled 
bank_sort_fill_df.sort_index()
```
